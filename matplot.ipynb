```python
import matplotlib.pyplot as pt
import numpy as np
y=np.array([10,20,30,40,50])
label=['apples','bananas','carrots','peas','ditch']
myexp=[1,1,1,1,1]
mycolors=['pink','purple','black','cyan','white']
pt.pie(y,labels=label,explode=myexp,colors=mycolors)
pt.show()
```
